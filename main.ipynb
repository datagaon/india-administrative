```python
import json
import pandas as pd
```

```python
villages = pd.read_csv('./files/A-3_MDDS_Release.csv')
villages
```

```python
columns_of_interest = ['State Code', 'District Code', 'Sub District Code', 'Level','Area Name']
villages[columns_of_interest].dropna(how='all').to_csv('./files/villages_with_administrative_hierarchy.csv', index=False)
villages = pd.read_csv('./files/villages_with_administrative_hierarchy.csv')
villages
```

```python
group_by_states = villages.groupby(by=['State Code','District Code'])
list(group_by_states[['Area Name','Level']])
```

```python
states = villages[villages['Level'] == 'STATE']
states[['State Code', 'Area Name']]
```

```python
districts = villages[villages['Level'] == 'DISTRICT']
districts[['State Code','District Code','Area Name']]
```

```python
code_to_states = {}
jsonable_shape = {}

def to_state(df):
  return code_to_states[df['State Code']]

for state in [row[1] for row in states[['State Code', 'Area Name']].iterrows()]:
  code_to_states[state['State Code']] = state['Area Name']
  jsonable_shape[to_state(state)] = {}

# print(code_to_states)
jsonable_shape
```

```python
code_to_district = {}

def to_district(df):
  return code_to_district[df['District Code']]

for district in [row[1] for row in districts[['State Code','District Code','Area Name']].iterrows()]:
  jsonable_shape[to_state(district)][district['Area Name']] = []
  code_to_district[district['District Code']] = district['Area Name']

# print(code_to_district)
jsonable_shape
```

```python
sub_districts = villages[villages['Level'] == 'SUB-DISTRICT']
for sd in [row[1] for row in sub_districts.iterrows()]:
  jsonable_shape[to_state(sd)][to_district(sd)].append(sd['Area Name'])

jsonable_shape
```

```python
def store_hierarchy(jsonable):
  with open("./outputs/hierarchy.json", "w") as f:
    json.dump(jsonable, f, indent=2)
```

```python
store_hierarchy(jsonable_shape)
```

Now let's separate out districts between J&K UT and Ladakh UT

```python
wkpd = pd.read_html('https://en.wikipedia.org/wiki/List_of_districts_of_Jammu_and_Kashmir')
jammu, kashmir = wkpd[0], wkpd[1]

jsonable_shape = None
if jsonable_shape is None:
  with open('./outputs/hierarchy.json', 'r') as f:
    jsonable_shape = json.load(f)

jk_state = jsonable_shape['JAMMU & KASHMIR']
```

```python
def headquarters_as_districts(df, districts):
  to_pick = [district[1]['Headquarters'][0] for district in df[['Headquarters']].iterrows()]
  return {k:v for k,v in districts if k in to_pick}
```

```python
kashmir_jsonable = headquarters_as_districts(kashmir, jk_state.items())
milale, pahije =(sorted(list(kashmir_jsonable.keys())), sorted(set([district[1]['Headquarters'][0] for district in kashmir[['Headquarters']].iterrows()])))
print(milale)
print(pahije)
[missing for missing in pahije if missing not in milale]
```

```python
jammu_districts = headquarters_as_districts(jammu, jk_state.items())
jammu_districts['Poonch'] = jk_state['Punch']
jammu_districts
```

```python
non_kashmir_districts = set(jammu_districts.keys())
non_kashmir_districts.update(['Leh(Ladakh)', 'Kargil'])
kashmir_district_names = list(set(jk_state.keys()) - non_kashmir_districts)
kashmir_districts = {k:v for k,v in jk_state.items() if k in kashmir_district_names}
kashmir_districts
```

```python
ladakh = {
  'LADAKH': {
    'Leh': jk_state['Leh(Ladakh)'], 
    'Kargil': jk_state['Kargil']
  }
}
print(ladakh)

jammu_and_kashmir = {
  'JAMMU & KASHMIR': {**jammu_districts, **kashmir_districts}
}
jammu_and_kashmir
```

```python
store_hierarchy({**jsonable_shape, **jammu_and_kashmir, **ladakh})
```
